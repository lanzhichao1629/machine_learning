### 数据的导入和预处理(回顾)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
datasets = pd.read_csv('studentscores.csv')
# print(datasets.isnull())             # 无缺失值
X = datasets.iloc[:,0:1].values    
#Y = datasets.iloc[:,-1].values
Y = datasets['Scores'].values

# 划分数据集
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state = 0)
# print(X_train,X_test,Y_train,Y_test)
```

### 训练线性模型并预测测试集

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_model = linear_regression.fit(X_train,Y_train)
linear_model.score(X_train,Y_train)
a = linear_model.coef_[0]                    # coef代表系数
b = linear_model.intercept_               # intercept代表截距
print("训练结果是：y = {:.3}x+{:.3}".format(a,b))      # 保留三位小数
# print(a,b)
predict = linear_model.predict(X_test)
print("测试集的测试结果为：{0}".format(predict))
```

### 可视化

```python
# 可视化训练集
plt.title("Train datasets", fontsize=24)
plt.xlabel("Hours", fontsize=14)
plt.ylabel("Scores", fontsize=14)
plt.scatter(X_train,Y_train)
plt.plot(X_train,9.91*X_train+2.02,'r')        # 训练的直线模型
plt.show()
```

```python
# 可视化测试集
plt.title("Test datasets", fontsize=24)
plt.xlabel("Hours", fontsize=14)
plt.ylabel("Scores", fontsize=14)
plt.scatter(X_test,Y_test)
plt.scatter(X_test,predict)
plt.show()
```
